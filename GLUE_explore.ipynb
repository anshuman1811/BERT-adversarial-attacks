```python
data_dir = '../glue_data/'
```

```python
import csv
def _read_tsv(input_file):
    """Reads a tab separated value file."""
    with open(input_file, "r", encoding="utf-8-sig") as f:
        return list(csv.reader(f, delimiter="\t"))
```

```python
mnli = pd.read_csv(os.path.join(data_dir,'MNLI/dev_matched.tsv'), delimiter='\t')
mnli_mm = pd.read_csv(os.path.join(data_dir,'MNLI/dev_mismatched.tsv'), delimiter='\t')
```
